```python
import Colorbin, Chorogrid
from Colorbin import Colorbin
from Chorogrid import Chorogrid
```

```python
mycolors = ['#E42217','#FF8040','#FFF380','#BDEDFF','#38ACEC','#2B65EC']
import pandas as pd
df = pd.read_csv('State-count-data.csv')
mybin = Colorbin(df['customer'], mycolors, proportional=True, decimals=None)
mybin.set_decimals(1)
mybin.fenceposts = [0, 10, 30, 50, 70, 90, 120 ]
mybin.recalc(fenceposts=False)
mybin.calc_complements(0.5, '#e0e0e0', '#101010')
mybin.fenceposts
```

```python
states = list(df.state)
colors_by_state = mybin.colors_out
font_colors_by_state = mybin.complements
legend_colors = mybin.colors_in
legend_labels = mybin.labels

for lst in ['states', 'colors_by_state', 'font_colors_by_state', 'legend_colors', 'legend_labels']:
    obj = eval(lst)
    print("{:>20}: len {:2}: {}...".format(lst, len(obj), obj[:3]))
```

```python
cg = Chorogrid('usa_states.csv', states, colors_by_state)
cg.set_title('Customer Base Count', font_dict={'font-size': 19})
cg.set_legend(legend_colors, legend_labels, title='Count Number')

cg.draw_hex(spacing_dict={'margin_right': 150, 'margin_bottom': 100}, font_colors=font_colors_by_state)
cg.done(show=True, save_filename = "CustomerBaseCount")
```
